```python
df_train['Date'] = pd.to_datetime(df_train['Date'], format='mixed') #'%d-%m-%Y'  #format='mixed'  #format='ISO8601'
df_train['TimeStamp'] = pd.to_datetime(df_train['TimeStamp'], format='mixed')
df_train
```

```python
df_train.fillna(0, inplace=True)
df_train
```

```python
# Extract year
df_train['year'] = df_train['TimeStamp'].dt.year
df_train['month'] = df_train['TimeStamp'].dt.month
df_train['day'] = df_train['TimeStamp'].dt.day
df_train['hour'] = df_train['TimeStamp'].dt.hour
df_train['dayofweek'] = df_train['TimeStamp'].dt.dayofweek
df_train['quarter'] = df_train['TimeStamp'].dt.quarter
```

```python
# Identify feature types
numerical_features = df_train.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df_train.select_dtypes(include=['object']).columns.tolist()
numerical_features, categorical_features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
# Define transformers
numerical_transformer = StandardScaler()  # or MinMaxScaler()
# categorical_transformer = LabelEncoder()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
for col_name in categorical_features:
    df_train[col_name] = categorical_transformer.fit_transform(df_train[col_name])

df_train
```

```python
df_train['TimeStamp'] = pd.to_datetime(df_train['TimeStamp'], format='mixed') #'%d-%m-%Y'  #format='mixed'  #format='ISO8601'
#drop the Date feature
df_train = df_train.drop(columns='Date', axis=1)
df_train
```

```python
#setting the 'TimeStamp' feature as index
df_train.set_index('TimeStamp', inplace=True)
```

```python
#check for the duplicate rows
duplicate_rows = df_train[df_train.index.duplicated()]
print("Duplicate Rows:\n", duplicate_rows)
```

```python
# Remove duplicates (keeping the first occurrence)
df_train = df_train.loc[~df_train.index.duplicated(keep='first')]
```

#### Missing holidays by checking the country

```python
# Convert strings to datetime
import holidays
from datetime import datetime

starting_year = df_reindexed.index.min().year
ending_year = df_reindexed.index.max().year
holiday_dates = df_reindexed[df_reindexed['Holiday']==1.0].index.to_list()

# function to check holidays for a country
def check_holidays_for_country(country_code, start_year, end_year, holiday_dates):
    country_holidays = holidays.CountryHoliday(country_code, years=range(start_year, end_year + 1))
    
    # list of holidays for the country in the given years
    matched_holidays = []
    for date in holiday_dates:
        if date in country_holidays:
            matched_holidays.append((date, country_holidays[date]))
    
    return matched_holidays

# List of country codes to check
country_codes = ['US', 'IN', 'GB', 'CA', 'AU', 'DE', 'FR', 'SG', 'PH', 'MY', 'IND']

Countris_listed = {}
count = 0
# Check holidays for each country and print matches
for country_code in country_codes:
    matched_holidays = check_holidays_for_country(country_code, starting_year, ending_year, holiday_dates)
    # print(country_code,len(matched_holidays))
    if len(matched_holidays)> count:
        print(country_code,len(matched_holidays))
        count = len(matched_holidays)
        max_ho_country = country_code
        print(f"\nMatched holidays for {country_code}:")
        for date, holiday_name in matched_holidays:
            print(f"{date.strftime('%Y-%m-%d')}: {holiday_name}")
```

```python
df_reindexed['Holiday'].value_counts()
```

```python
# Apply the function to fill missing holidays in the 'Holiday' column

us_country_holidays = holidays.US(years=range(starting_year, ending_year + 1))

def fill_missing_holidays(row):
    timestamp = row.name  # The index should be the timestamp, so use row.name
    # Check if the 'Holiday' column is NaN
    if pd.notna(row['Holiday']):
        return row['Holiday']
    else:
        # Check if the timestamp is in the US holidays dictionary
        if timestamp in us_country_holidays:
            return 1.0  # It is a holiday
        return 0.0  # Not a holiday


df_reindexed['Holiday'] = df_reindexed.apply(lambda row: fill_missing_holidays(row), axis=1)
```
